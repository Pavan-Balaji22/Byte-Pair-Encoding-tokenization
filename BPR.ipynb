```python
def get_counts(byteData):
    counts = {}
    for i,j in zip(byteData,byteData[1:]):
        counts[i,j]=counts.get((i,j),0)+1
    return counts

def get_indices(keys,byteData):
    indices = {}
    n = 0
    for i,j in enumerate(zip(byteData,byteData[1:])):
        if j[0] == keys[0] and j[1] == keys[1]:
            indices[n] =(i,i+1)
            n+= 1
    return indices 
    
```

```python
bp_counts = get_counts(data_bytes)
```

```python
req_key = max(bp_counts.items(),key=lambda x:x[1])[0]
```

```python
get_indices(req_key,data_bytes)
```
